
Q1. What is a parameter?

Ans:-
A parameter is a value learned by a Machine Learning model during training. It controls how input features are combined to produce output. Parameters are updated to reduce error.

---

Q2. What is correlation?

Ans:-
Correlation measures the strength and direction of the relationship between two variables. It shows how variables change together. Its value ranges from -1 to +1.

---

Q3. What does negative correlation mean?

Ans:-
Negative correlation means that as one variable increases, the other decreases. Both variables move in opposite directions. Example: price increase and demand decrease.

---

Q4. Define Machine Learning. What are the main components?

Ans:-
Machine Learning is a field of AI where systems learn from data without explicit programming. Main components include data, features, model, loss function, optimizer, and evaluation metrics.

---

Q5. How does loss value help in determining whether the model is good or not?

Ans:-
Loss value measures the difference between predicted and actual values. A lower loss indicates better model performance. High loss means poor predictions.

Q6. What are continuous and categorical variables?

Ans:-
Continuous variables are numerical and can take infinite values, such as height or salary. Categorical variables represent groups or labels, such as gender or city.

Q7. How do we handle categorical variables? What are common techniques?

Ans:-
Categorical variables are converted into numerical form. Common techniques include Label Encoding, One-Hot Encoding, and Ordinal Encoding.

Q8. What do you mean by training and testing a dataset?

Ans:-
Training data is used to teach the model patterns. Testing data is used to evaluate the model on unseen data. This checks model accuracy and generalization.

A test set is a part of the dataset kept separate from training data. It is used to evaluate the final performance of the trained model.

Q16. Difference between correlation and causation with example.

**Correlation:**

* Shows a relationship between two variables

* One variable does not directly cause the other

* Example: Ice cream sales and drowning cases increase together

**Causation:**

* Shows a direct cause-and-effect relationship

* One variable directly affects another

* Example: Smoking causes lung cancer

Q17. What is an Optimizer? Types with examples.

Ans:-
**Optimizer:**
An optimizer is an algorithm used to update the model parameters (weights and bias) in order to minimize the loss function and improve model accuracy.

**Types of Optimizers:**

* Gradient Descent: Updates parameters using the entire dataset.
Example: Used in simple linear regression.

* Stochastic Gradient Descent (SGD): Updates parameters using one data point at a time.
Example: Used in large datasets for faster learning.

* Adam Optimizer: Combines advantages of SGD and RMSprop for faster convergence.
Example: Commonly used in deep learning models.

* RMSprop: Adjusts learning rate based on recent gradients.
Example: Used in neural networks with noisy data.
